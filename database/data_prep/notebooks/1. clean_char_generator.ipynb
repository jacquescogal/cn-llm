```python
import pandas as pd

raw_char_db = pd.read_csv("./dataset/raw/hsk3_all_characters.csv")
raw_char_db.head()
```

```python
# clean the raw dataset
clean_char_db = raw_char_db.drop(columns=["Unnamed: 0"])
clean_char_db.columns = clean_char_db.iloc[1]
clean_char_db.drop([0,1,2], inplace=True)
clean_char_db = clean_char_db.iloc[:, [1,9,10,13]]
clean_char_db.columns = ["hsk_level", "char_hanzi", "pinyin", "meaning"]
clean_char_db = clean_char_db[["char_hanzi", "pinyin", "meaning", "hsk_level"]]
clean_char_db.reset_index(drop=True, inplace=True)
clean_char_db.head()
```

```python
# remove rows with nan values (they are subheaders)
for i in clean_char_db.iterrows():
    # check if nan and remove
    if pd.isnull(i[1]["char_hanzi"]):
        clean_char_db.drop(i[0], inplace=True)
clean_char_db.reset_index(drop=True, inplace=True)
```

```python
clean_char_db.index += 1
clean_char_db.to_csv("./dataset/clean/hsk3_all_characters.csv", index=True, index_label="char_id")
```
