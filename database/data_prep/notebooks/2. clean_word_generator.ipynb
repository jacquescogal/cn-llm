```python
import pandas as pd

raw_word_db = pd.read_csv("../dataset/raw/hsk3_all_words.csv")
raw_word_db.head()
```

```python
# clean the raw dataset
clean_word_db = raw_word_db.drop(columns=["Unnamed: 0"])
clean_word_db.columns = clean_word_db.iloc[1]
clean_word_db.head()
clean_word_db.drop([0,1,2], inplace=True)
clean_word_db = clean_word_db.iloc[:, [1,2,3,8]]
clean_word_db.columns = ["word_hanzi", "pinyin", "meaning", "hsk_level"]
clean_word_db.reset_index(drop=True, inplace=True)
clean_word_db.head()
```

```python
# remove rows with nan values (they are subheaders) and hsk_level > 6
for tup in clean_word_db.itertuples():
    if pd.isnull(tup.word_hanzi) or tup.hsk_level.isdigit() is False:
        clean_word_db.drop(tup.Index, inplace=True)
clean_word_db.reset_index(drop=True, inplace=True)
```

```python
print(len(clean_word_db))
clean_word_db.index += 1
clean_word_db.to_csv("../dataset/clean/hsk3_all_words.csv", index=True, index_label="word_id")
```
